```python
import pandas as pd
import requests
import time
import numpy as np
from bs4 import BeautifulSoup

dataframe=pd.read_csv("../final data/college maping.csv")
dataframe.head()
```

```python
length=dataframe.shape
dataframe["Institute dict"]=np.nan
```

```python
for i in range(length[0]):
    url="http://www.dtemaharashtra.gov.in/frmInstituteSummary.aspx?InstituteCode={}".format(dataframe["Institute Code"].iloc[i])
    r=requests.get(url)
    html_conent=r.content
    soup=BeautifulSoup(html_conent,'html.parser')
    classs=soup.find(id="ctl00_ContentPlaceHolder1_lblDistrict")
    dataframe["Institute dict"].iloc[i]=classs.text
    time.sleep(3)
```

```python
dataframe.head(100)
```

```python
dataframe.drop(["Unnamed: 0","Institute Name"],axis=1,inplace=True)
```

```python
dataframe.to_csv("../final data/college town.csv")
```
